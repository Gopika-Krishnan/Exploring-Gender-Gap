# Examining the Gender Gap by Region


<p>We can see from the following visualization that employement in general seems to be greater in Northern America followed by Southern Asia. From all regions, we see the gender gap in employement, with the male population leading.</p>
  

```python
print(db['region'].unique(), "\nTotal=",db['region'].nunique())#number of regions and regions in the dataframe
```

```python
f=[]
m=[]
region = db.loc[db['region']!='empty']['region'].unique() #empty is removed

#making lists f and m counting employment numbers by region
for r in region:
    f.append((db.loc[(db['region']==r)&(db['gender']=='female')]['count'].sum())/1000000)
    m.append((db.loc[(db['region']==r)&(db['gender']=='male')]['count'].sum())/1000000)
x = np.arange(len(region))  # the label locations
width = 0.35  # the width of the bars
```

```python
fig,ax = plt.subplots(figsize=(16,10))

rects1 = ax.bar(x - width/2, m, width, label='Men',color='#AFAFF5')
rects2 = ax.bar(x + width/2, f, width, label='Women',color='#EFAFB5')

ax.legend(['Male','Female'])
plt.ticklabel_format(style='plain',axis='y')

ax.set_xlabel('Region',fontsize = 20)
ax.set_ylabel('Number of Employees (in millions)',fontsize = 20)
ax.set_title('Number of People employed - Region wise (2012-2020)',fontsize = 25)
ax.set_xticklabels(region)
ax.set_xticks(x)
ax.figure.autofmt_xdate()
```


<p>In the next visual of this section, we analyse the percentage makeup of male/female workforce. It can be seen that in all cases, the percentage of male employees is seen to be greater than 50%.</p>

```python
#new dataframe with male and female workforce per region
df2 = pd.DataFrame(list(zip(region,f, m)), columns =['Region','Female', 'Male'])
df2
```

```python
#compute percentages
df2['totals'] = df2['Female']+df2['Male']
df2['femalepercentages'] = (df2['Female']*100)/df2['totals']
df2['malepercentages'] =  (df2['Male']*100)/df2['totals']
df2
```

```python
malepercentages=df2['malepercentages'].sort_values(ascending=True)
femalepercentages=df2['femalepercentages'].sort_values(ascending=False)
length = range(len(femalepercentages))

#Plots the bar chart
fig = plt.figure(figsize=(7, 7))
sns.set(font_scale=1.5)
p1 = plt.barh(length, malepercentages.values, 0.55, label='Male', color='#AFAFF5')
p2 = plt.barh(length, femalepercentages, 0.55, left=malepercentages, color='#EFAFB5', label='Female')



labels = malepercentages.index
plt.yticks(range(len(malepercentages)), region)
plt.xticks([0, 25, 50, 75, 100], ['0 %', '25 %', '50 %', '75 %', '100 %'])
plt.xlabel('Percentage of Male Workforce')
plt.ylabel('Region')
plt.title('Gender Representation by Region',fontsize=20)
plt.axvline(x=50, color='k', linestyle='dashed', linewidth=1)
plt.legend(bbox_to_anchor=(0, 1, 1.5, 0))
plt.show()
```

# Examining the Gender Gap by Sector


<p>The gender makeup of the workforce is examined by sector in this section. In all industries except the health care sector, it showed that the number of people employed in that particular industry was mostly male.</p>

```python
sector = db['sector'].unique()
my_range=range(1,len(sector)+1)
f1=[]
for r in sector:
    f1.append(db.loc[(db['sector']==r)&(db['gender']=='female')]['count'].sum()/1000000)
    

m1=[]
for r in sector:
    m1.append(db.loc[(db['sector']==r)&(db['gender']=='male')]['count'].sum()/1000000)
    
fig,ax = plt.subplots(figsize=(14,9))  
plt.hlines(y=range(1,len(sector)+1),xmin=min(f1,m1), xmax=max(f1,m1), color='gray', alpha=0.4)
plt.scatter(f1, y=range(1,len(sector)+1),color='deeppink', alpha=1, label='female')
plt.scatter(m1, y=range(1,len(sector)+1),color='blue', alpha=0.8 , label='male')
plt.legend()
plt.yticks(my_range, sector)
plt.xlabel('Number of people employed (in millions)')
plt.ylabel('Sector')
plt.ticklabel_format(style='plain', axis='x')
plt.title('Gender Representation by Sector',fontsize=25)
```

```python
sector = db['sector'].unique()
my_range=range(1,len(sector)+1)
b1=[]
for r in sector:
    b1.append(db.loc[(db['sector']==r)&(db['gender']=='female')&(db['year']=='2020')]['count'].sum()/1000000)
    

b2=[]
for r in sector:
    b2.append(db.loc[(db['sector']==r)&(db['gender']=='male')&(db['year']=='2020')]['count'].sum()/1000000)
    
fig,ax = plt.subplots(figsize=(14,9))  
plt.hlines(y=range(1,len(sector)+1),xmin=min(b1,b2), xmax=max(b1,b2), color='gray', alpha=0.4)
plt.scatter(b1, y=range(1,len(sector)+1),color='deeppink', alpha=1, label='female')
plt.scatter(b2, y=range(1,len(sector)+1),color='blue', alpha=0.8 , label='male')
plt.legend()
plt.yticks(my_range, sector)
plt.xlabel('Number of people employed (in millions)')
plt.ylabel('Sector')
plt.ticklabel_format(style='plain', axis='x')
plt.title('Gender Representation by Sector in 2020',fontsize=25)
```
